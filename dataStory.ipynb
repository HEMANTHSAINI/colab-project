```python
import pandas as pd
import plotly.express as px
df=pd.read_csv("savings_data_final.csv")
fig=px.scatter(df,y="quant_saved",color="rem_any")
fig.show()
```

```python
import csv
import plotly.graph_objects as go
with open('savings_data_final.csv',newline='')as f:
  reader=csv.reader(f)
  savings_data=list(reader)
savings_data.pop(0)
total_entries=len(savings_data)
total_rem=0
for data in savings_data:
  if int(data[3])==1:
    total_rem+=1
fig=go.Figure(go.Bar(x=["reminded","notreminded"],y=[total_rem,(total_entries-total_rem)]))
fig.show()
```

```python
rem=[]
not_rem=[]
for data in savings_data:
  if int(data[3])==1:
    rem.append(float(data[0]))
  else:
    not_rem.append(float(data[0]))
print(f"mean of rem - {statistics.mean(rem)}")
print(f"median of rem - {statistics.median(rem)}")
print(f"mode of rem - {statistics.mode(rem)}")

print(f"mean of no rem - {statistics.mean(not_rem)}")
print(f"median of no rem - {statistics.median(not_rem)}")
print(f"mode of no rem - {statistics.mode(not_rem)}")
```

```python
import statistics
savings=[]
for data in savings_data:
  savings.append(float(data[0]))
print(f"mean - {statistics.mean(savings)}")
print(f"median - {statistics.median(savings)}")
print(f"mode - {statistics.mode(savings)}")
```

```python
print(f"sd of all data - {statistics.stdev(savings)}")
print(f"sd of people who were reminded - {statistics.stdev(rem)}")
print(f"sd of people who were not reminded - {statistics.stdev(not_rem)}")
```

```python
import numpy as np
age=[]
savings_per_age=[]
for data in savings_data:
  if float (data[5])!=0:
    age.append(float(data[5]))
    savings_per_age.append(float(data[0]))
correlation=np.corrcoef(age,savings_per_age)
print(f"correlation between age and there savings={correlation[0,1]}")
```

```python
import plotly.figure_factory as ff
fig=ff.create_distplot([df["quant_saved"].tolist()],["Savings"],show_hist=False)
fig.show()
```

```python
q1=df['quant_saved'].quantile(0.25)
q3=df['quant_saved'].quantile(0.75)
iqr=q3-q1
lowerwhisker=q1-1.5*iqr
upperwhisker=q3+1.5*iqr
print(q1,q3,iqr)
```

```python
newdf=df[df['quant_saved']<upperwhisker]
```

```python
allsavings=newdf['quant_saved'].tolist()
mean=statistics.mean(allsavings)
median=statistics.median(allsavings)
mode=statistics.mode(allsavings)
sd=statistics.stdev(allsavings)
print(mean,median,mode,sd)
```

```python
fig=ff.create_distplot([newdf['quant_saved'].tolist()],['savings'],show_hist=False)
fig.show()
```

```python
import random

samplingmeanlist=[]
for i in range(1000):
  templist=[]
  for j in range(100):
    templist.append(random.choice(allsavings))
  samplingmeanlist.append(statistics.mean(templist))

mean=statistics.mean(samplingmeanlist)
sd=statistics.stdev(samplingmeanlist)
print(mean,sd)
```

```python
fig=ff.create_distplot([samplingmeanlist],['sampledsavings'],show_hist=False)
fig.show()
```

```python
tempdf=newdf[newdf.age!=0]
age=tempdf['age'].tolist()
savings=tempdf['quant_saved'].tolist()
c=np.corrcoef(age,savings)
print(c)
```

```python
remdf=newdf.loc[newdf['rem_any']==1]
notremdf=newdf.loc[newdf['rem_any']==0]
print(remdf.head())
print(notremdf.head())
```

```python
remsavings=remdf['quant_saved'].tolist()
remsamplingmeanlist=[]
for i in range(1000):
  templist=[]
  for j in range(100):
    templist.append(random.choice(remsavings))
  remsamplingmeanlist.append(statistics.mean(templist))

remmean=statistics.mean(remsamplingmeanlist)
remsd=statistics.stdev(remsamplingmeanlist)
print(remmean,remsd)
```

```python
notremsavings=notremdf['quant_saved'].tolist()
notremsamplingmeanlist=[]
for i in range(1000):
  templist=[]
  for j in range(100):
    templist.append(random.choice(notremsavings))
  notremsamplingmeanlist.append(statistics.mean(templist))

notremmean=statistics.mean(notremsamplingmeanlist)
notremsd=statistics.stdev(notremsamplingmeanlist)
print(notremmean,notremsd)
```

```python
zscore=(remmean-notremmean)/notremsd
print(zscore)
```
